## Soluciones:


*   1) Voy a eliminar todos los renglones, si tiene al menos un dato pérdido 
*   2) Voy a eliminar por columnas
*   3) Voy a eliminar renglones si todos los valores son pérdidos
*   4) Si existe al menos "n" datos reales, no voy a eliminar estos renglones
*   5) Reemplazar con promedio
*   6) Reemplzar con la mediana
*   7) Reemplazar con la moda





```python
#1
ndf = df.copy()
#inplace lo que hace es que elimina los datos pérdidos
ndf.dropna(inplace = True)
ndf
#Pero vemos que elimina muy pocos
```

```python
#2
ndf2 = df.copy()

ndf2.dropna(inplace = True,
           axis = 1. #aplicar por todas las columnas
           )
ndf2
#Básicamente elimina 23 de las 25 columnas, es una estrategia muy mala en este caso, dado que hay muchos datos pérdidos
```

```python
#3
ndf3 = df.copy()
ndf3.dropna(how = 'all', #elimina los datos si todos son pérdidos con el all
            inplace = True)
ndf3
# En este caso tampoco es tan efectivo ya que no eliminó ninguno
```

```python
#4
ndf4 = df.copy()
ndf4.dropna(thresh = 23, #elimina los renglones que tengan menos de "n" cantidad de elementos
            inplace = True)
ndf4
# En este caso tampoco es tan efectivo ya que no eliminó ninguno
```

```python
#4.1 para que aplique en columnas le ponemos el axis = 1 
ndf4 = df.copy()
ndf4.dropna(thresh =25000, #elimina los renglones que tengan menos de "n" cantidad de elementos
            inplace = True,
            axis = 1)
ndf4
# En este caso tampoco es tan efectivo ya que no eliminó ninguno
```

```python
#5 reemplzar con el promedio los datos pérdidos
ndf5 = df.copy()
ndf5['X5'].fillna( value = ndf5['X5'].mean(),
                  inplace = True)
ndf5
```

```python
#6 reemplzar con el la mediana los datos pérdidos
ndf6 = df.copy()
ndf6['X5'].fillna( value = ndf6['X5'].median(),
                  inplace = True)
ndf6
```

```python
#7 Reemplzar con la moda
# Primero debemos de ver cuál es la moda
df['X5'].mode()
```

```python
# En nuestro caso es el número 29 y procedemos a guardarlo en una variable
oc = df['X5'].mode()[0]
ndf7 =  df.copy()
ndf7['X5'].fillna(value = oc,
                       inplace = True)
ndf7
```

## ¿Se eliminaron o reemplazaron datos nulos? ¿Qué se hizo y por qué?


Se aplicaron todas las soluciones propuestas en la actividad ejemplo.

## ¿Es necesario ordenar los datos para el análisis? Sí / No / ¿Por qué?


Llegamos a la conclusión de que el orden de los datos no afecta de ninguna manera el análisis de los datos. Las propiedades que se utilizan para el análisis como el promedio, la moda, la varianza y en general toda la estadística no depende de que los datos estén ordenados de cierta manera. El único parámetro que se ve afectado por el orden es la mediana, sin embargo, no tiene gran relevancia para los demás análsis.

## ¿Qué ajustes se realizaron en el proceso de limpieza de datos (agregar, integrar, eliminar, modificar registros (filas), cambiar atributos (columnas)?


Se le realizaron varios ajustes a la base de datos para tenerla lo más limpia posible. En específico se realizaron siete ajustes diferentes modificaciones:
- Se eliminaron todos los renglones que tuvieran al menos un dato faltante dentro de la base de datos.
- Se eliminaron todas las columnas que tuvieran al menos un dato faltante dentro de la base de datos.
- Se eliminaron todos los renglones que tuvieran todos los datos faltantes.
- Se puso un límite "n" de datos faltantes para determinar qué renglones debían ser eliminados.
- Se reemplazaron los datos faltantes con el promedio de los datos existentes de la base de datos.
- Se reemplazaron los datos faltantes con la mediana de los datos existentes de la base de datos.
